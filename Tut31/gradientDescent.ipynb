```python
# Gradient descent (also often called steepest descent) is a first-order iterative optimization algorithm 
# for finding a local minimum of a differentiable function. The idea is to take repeated steps in the opposite 
# direction of the gradient (or approximate gradient) of the function at the current point, because this is the 
# direction of steepest descent. Conversely, stepping in the direction of the gradient will lead to a local 
# maximum of that function; the procedure is then known as gradient ascent.
```

```python
# L -> b^2
# bnew = bold - slope
# bnew = bold - (eta)*slope , where eta = learning rate
# when to stop:
# bnew ~= bold (not much improvement any more)
# no. of interations(epochs) = 1000 etc.
```

```python
from sklearn.datasets import make_regression 
import numpy as np
import matplotlib.pyplot as plt
```

```python
X, y = make_regression(n_samples = 4, n_features= 1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
plt.scatter(X,y)
```

```python
# Lets apply Gradient Descent assuming slope is constant m = 78.35
# and let's assume the starting value for intercept b = 0
y_pred = ((78.35 * X) + 100).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred,color='#00a65a',label='b = 0')
plt.legend()
plt.show()
```

```python

m = 78.35
b = 100

loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
# Lets take learning rate = 0.1
lr = 0.1

step_size = loss_slope*lr
step_size
```

```python
# Calculating the new intercept
b = b - step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
# Iteration 2
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,y_pred2,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred1,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
# Iteration 3
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred3 = ((78.35 * X) + b).reshape(4)

plt.figure(figsize=(15,15))
plt.scatter(X,y)
plt.plot(X,y_pred3,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred2,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred1,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
b= -500
m = 78.35
epochs = 10
for i in range(epochs):
    loss_slope = -2 * np.sum(y - m*X.ravel() - b)
    b = b - (lr*loss_slope)
    y_pred = m*X +b
    plt.plot(X, y_pred)

plt.scatter(X,y)
```
